```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
total_players = len(purchase_data['SN'].unique())

total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data['Item ID'].unique())

average_price = purchase_data['Price'].mean()

number_purchases = purchase_data['Price'].count()

total_revenue = purchase_data['Price'].sum()

summary = pd.DataFrame({
    'Number of Unique Items': [unique_items],
    'Average Price': average_price,
    'Number of Purchases': number_purchases,
    'Total Revenue': total_revenue})

summary['Average Price'] = summary['Average Price'].map("${:.2f}".format)
summary['Total Revenue'] = summary['Total Revenue'].map("${:,.2f}".format)
summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
new_data = purchase_data.drop_duplicates('SN')

gender_count = new_data['Gender'].value_counts()

percent = (gender_count / total_players) * 100

gender_counts = pd.DataFrame({
    'Total Count': gender_count,
    'Percentage of Players': percent})

gender_counts['Percentage of Players'] = gender_counts['Percentage of Players'].map("{:.2f}%".format)

gender_counts
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_df = purchase_data.groupby('Gender')

#calculations
genders = gender_df.mean()
genders['Total Purchase Value'] = gender_df['Price'].sum()
genders['Purchase Count'] = gender_df['Purchase ID'].count()
genders['Average Total Purchase Per Person'] = genders['Total Purchase Value'] / gender_counts['Total Count']

#formatting
genders = genders.rename(columns ={'Price': 'Average Purchase Price'})
genders['Average Purchase Price'] = genders['Average Purchase Price'].map("${:.2f}".format)
genders['Total Purchase Value'] = genders['Total Purchase Value'].map("${:.2f}".format)
genders['Average Total Purchase Per Person'] = genders['Average Total Purchase Per Person'].map("${:.2f}".format)

#getting the correct columns
genders[['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Average Total Purchase Per Person']]
```

* Create a summary data frame to hold the results

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
new_data['Age Group'] = pd.cut(new_data["Age"], bins, labels=group_names)
age_data = new_data.groupby('Age Group')
ages = age_data.count()
ages['Percentage of Players'] = (age_data['Purchase ID'].count() / total_players) * 100
ages = ages.rename(columns ={'Purchase ID': 'Total Count'})
age_summary = ages[['Total Count', 'Percentage of Players']]



age_summary['Percentage of Players'] = age_summary['Percentage of Players'].map("{:.2f}%".format)
age_summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data['Age Group'] = pd.cut(purchase_data["Age"], bins, labels=group_names)

#calculations
age_df = purchase_data.groupby('Age Group')
age_groups = age_df.count()
age_groups['Average Purchase Price'] = age_df['Price'].mean()
age_groups['Total Purchase Value'] = age_df['Price'].sum()
age_groups['Average Total Purchase Per Person'] = age_groups['Total Purchase Value'] / age_summary['Total Count']

#formatting
age_groups = age_groups.rename(columns = {'Purchase ID': 'Purchase Count'})
age_groups['Average Purchase Price'] = age_groups['Average Purchase Price'].map("${:.2f}".format)
age_groups['Average Total Purchase Per Person'] = age_groups['Average Total Purchase Per Person'].map("${:.2f}".format)
age_groups['Total Purchase Value'] = age_groups['Total Purchase Value'].map("${:.2f}".format)
age_groups_summary = age_groups[['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Average Total Purchase Per Person']]
age_groups_summary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#calculations
spenders = purchase_data.groupby('SN')
top_spenders = spenders.count()
top_spenders['Total Purchase Value'] = spenders['Price'].sum()
top_spenders['Average Purchase Price'] = top_spenders['Total Purchase Value'] / top_spenders['Purchase ID']
top_spenders = top_spenders.sort_values('Total Purchase Value', ascending=False)

#formatting
top_spenders = top_spenders.rename(columns ={'Purchase ID': 'Purchase Count'})
top_spenders['Total Purchase Value'] = top_spenders['Total Purchase Value'].map("${:.2f}".format)
top_spenders['Average Purchase Price'] = top_spenders['Average Purchase Price'].map("${:.2f}".format)
top_spenders_summary = top_spenders[['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']]
top_spenders_summary.head()
```
